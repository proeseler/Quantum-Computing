# Import Libraries

```python
from qiskit import QuantumCircuit, Aer, transpile
from qiskit.tools.visualization import plot_histogram
```

# Build quantum ciruit for Bell state

```python
def hello_quantum():
    # Define a quantum circuit with 2 qubits
    qc = QuantumCircuit(2)

    # Apply a Hadamard gate to the first qubit
    qc.h(0)

    # Apply a controlled-X (CNOT) gate using the first qubit as control and the second as target
    qc.cx(0, 1)
    qc.measure_all()

    # Use Aer's qasm_simulator
    simulator = Aer.get_backend("aer_simulator")

    # Compile the circuit for the simulator
    compiled_circuit = transpile(qc, simulator)

    # Run the compiled circuit on the simulator
    job = simulator.run(compiled_circuit, shots=10000)

    # Get the results
    result = job.result()

    # Get the counts of each outcome
    counts = result.get_counts(compiled_circuit)

    # Return the results
    return counts
```

# Execute circuit

```python
counts = hello_quantum()
# For this simple program, it should show roughly 50% |00> and 50% |11> if everything worked correctly.
plot_histogram(counts, title='Bell state counts')
```
